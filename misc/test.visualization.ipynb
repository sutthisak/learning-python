# Data visualization in Python
## Part 1: matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

HR dataser from here https://www.kaggle.com/ludobenistant/hr-analytics

```python
df = pd.read_csv("../data/HR_comma_sep.csv")
df.head()
```

Plot a simple histogram of the “number_project” column.

```python
df.number_project.plot('hist')
plt.show()
```

Plot histograms by a categorical variable.

```python
df.Work_accident.hist(by=df.left)
plt.show()
```

Plot a simple box plot. A box plot gives you a lot of statistically useful information such as the the 1st, 2nd and 3rd quartiles as well as outliers.

```python
df.boxplot(column = 'satisfaction_level')
plt.show()
```

Plot box plots by categorizing by a particular column ‘left’ had two levels – ‘0’ and ‘1’.

```python
df.boxplot(column = 'satisfaction_level', by = 'left')
plt.show()
```

Next is scatter plot. Scatterplots show you the relationship between two variables X and Y.

```python
df.plot(kind = 'scatter', x = 'average_montly_hours', y = 'satisfaction_level', xlim = (100,200))
plt.show()
```

 If we want to use the ‘ggplot’ style for our plots we can use the code shown below.

```python
plt.style.use('ggplot')
df.boxplot(column = 'satisfaction_level')
plt.show()
```

## PART TWO: Seaborn
Seaborn is fundamentally a layer over matplotlib and offers you a wide array of options when it comes to statistical plots compared to matplotlib.

We can plot a scatterplot coupled with a linear regression line of best fit by using the code shown below

The ‘hue’ argument was used to categorize it by a categorical variable while the ‘palette’ argument was used to give colors to the plot using the ‘Set1’ palette.

```python
import seaborn as sns
```

```python
sns.lmplot(x = 'average_montly_hours', y = 'satisfaction_level', data = df, hue = 'left', palette = 'Set1')
plt.show()
```

Stripplots are useful to visualize relationships between continuous and categorical variables.

```python
sns.stripplot(x = 'left', y = 'average_montly_hours', data = df)
plt.show()
```

Pairplots are a great way to visualize the relationship between all the numerical variables in your dataset.

```python
sns.pairplot(df, hue = "left")
plt.show()
```

Joint plots are useful because it helps you visualize the relationship between two variables X and Y and it gives you the individual histogram distributions of the two variables.

```python
sns.jointplot(x = "satisfaction_level", y = "average_montly_hours", data = df)
plt.show()
```

The most useful aspect about joint plots are the fact that you can use an argument called ‘kind’ to specify the kind of plot you want. For example if we use kind = ‘kde’ we get a density plot as shown below

```python
sns.jointplot(x = "satisfaction_level", y = "last_evaluation", data = df, kind = 'kde')
plt.show()
```
