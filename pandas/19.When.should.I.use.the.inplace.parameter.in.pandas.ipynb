```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo.shape
```

ยกตัวอย่างเช่นถ้าเราไม่สนใจคอลัมน์ City

```python
ufo.drop('City', axis=1)
```

```python
ufo.head()
```

จะเห็นว่า drop method ไม่ได้ลบคอลัมน์ที่เราต้องการออกไปจริงๆเพราะ พารามิเตอร์โดยปริยาย inplace=False ดังนั้นเราต้องกำหนดให้เป็น True ตั้งแต่ตอนเรียก drop method เพื่อให้ลบจริงๆ

```python
ufo.drop('City', axis=1, inplace=True)
ufo.head()
```

การที่พารามิเตอร์ inplace=False เป็นค่าโดยปริยายทำให้เราสามารถตรวจสอบผลกระทบได้ก่อนการกระทำการใดๆลงไป เช่นด้านล่างนี้เราสามารถทำการเปรียบเทียบว่าถ้าเราลบแถวที่มีคอลัมน์ไดคอลัมน์หนึ่งมีค่า Nan ออกไปจะเหลือข้อมูลให้ใช้งานได้อีกเท่าใดเป็นต้น

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

ในทางปฏิบัตินั้นเราสามารถสร้างตัวแปรมารองรับกับ method ที่มีพารามิเตอร์ inplace=False ได้ แต่ถ้าใช้ชื่อตัวแปรเดิมก็ให้ผลเหมือนกับ inplace=True นั่นเอง

```python
ufo = ufo.drop('Time', axis=1)
ufo.head()
```
