รื้อพื้นความจำเรื่อง loc และ iloc เกี่ยวกับการทำ slice numpy array โดย iloc จะไม่รวม index ด้านขวา

เรื่องของการสุ่ม ยกตัวอย่างสุ่ม 3 จำนวน

```python
ufo.sample(n=3)
```

สุ่มอีกครั้งเป็นจำนวน 3 จำนวน

```python
ufo.sample(n=3)
```

ถ้าต้องการสุ่มให้ได้ค่าเดิมเหมือนการสุ่มก่อนหน้า เช่นการทำ reproduce code ให้กำหนดพารามิเตอร์ random_state=ตัวเลข เช่นในที่นี้กำหนดตัวเลข=1 ต่อไปไม่ว่าจะสุ่มกี่ครั้งที่ random_state=1 ก็จะได้ค่าเดิมทุกครั้ง

```python
ufo.sample(n=3, random_state=1)
```

สุ่มอีกครั้งโดยมีค่า random_state=1 เหมือนเดิม

```python
ufo.sample(n=3, random_state=1)
```

ต่อมาคือวิธีการสุ่มให้ได้จำนวน 75% ของข้อมูล

```python
ufo.sample(frac=0.75, random_state=99)
```

การนำไปใช้งานเช่นการทำ Machine Learning นั้นเรามักจะทำการสุ่มข้อมูลการเทรนที่ 75% และข้อมูลเทส 25% ได้ง่ายๆ ดังนี้

```python
train = ufo.sample(frac=0.75, random_state=99)
test = ufo.loc[~ufo.index.isin(train.index), :]
```
