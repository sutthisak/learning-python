```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head(3)
```

loc ใช้สำหรับเลือกแถวและคอลัมน์ที่จะแสดงโดยตัวอย่างนี้เลือกแถวที่ 0 แสดงทุกคอลัมน์

```python
ufo.loc[0, :]
```

ต่อมาเลือกแถวที่ 0 1 2 และแสดงที่คอลัมน์

```python
ufo.loc[[0, 1, 2], :]
```

หรือแบบนี้ก็ได้ผลเช่นเดียวกัน

```python
ufo.loc[0:2, :]
```

เราสามารถเลือกทุกแถวเฉพาะคอลัมน์ City ได้ดังนี้

```python
ufo.loc[:, 'City']
```

เราสามารถเลือกทุกแถวเฉพาะคอลัมน์ City กับ State ได้ดังนี้

```python
ufo.loc[:, ['City', 'State']]
```

เรายังสามารถเลือกทุกแถวและทุกคอลัมน์ตั้งแต่ City ถึง State ได้ดังนี้

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.head(3).drop('Time', axis=1)
```

ลองใช้วิธีเดิมที่เคยทำงมาก่อนหน้านี้ในการเลือกข้อมูลที่จะแสดง

```python
ufo[ufo.City == 'Oakland']
```

ถ้าเลือกใช้ loc ก็ได้ผลอย่างเดียวกัน

```python
ufo.loc[ufo.City == 'Oakland', :]
```

แต่เราสามารถเลือกคอลัน์ที่เราต้องการได้ดังนี้

```python
ufo.loc[ufo.City == 'Oakland', ['State', 'Time']]
```

ซึ่งถ้าใช้วิธีเดิมเราจะเลือกได้เพียงคอลัมน์เดียวดังนี้

```python
ufo[ufo.City == 'Oakland'].State
```

หลังจากพอจะรู้จะ loc แล้ว มาทำความรู้จักกับ iloc กันต่อ โดย iloc เป็นการเลือกแถวและคอลัมน์โดยใช้ตำแหน่งจากตัวเลข ดังตัวอย่าง

```python
ufo.iloc[:,[0, 3]]
```

```python
ufo.iloc[:, 0:4]
```

```python
ufo.columns
```

ข้อสังเกตุจากการรัน ufo.iloc[:, 0:4] พบว่าคอลัมน์ที่แสดงคือ 0,1,2,3 ไม่รวมคอลัมน์ที่ 4

```python
ufo.iloc[0:3, :]
```

*** จากการรันด้านบนพบว่าแถวที่แสดงจะไม่แสดงแถวที่ 3 ซึ่งต่างจากการใช้ loc ที่จะแสดงแถวที่ 3 ด้วยดังนี้ ***

```python
ufo.loc[0:3, :]
```
